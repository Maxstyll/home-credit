# Origem dos dados

https://www.kaggle.com/c/home-credit-default-risk


Muitas pessoas lutam para obter empréstimos devido a históricos de crédito insuficientes ou inexistentes. E, infelizmente, essa população costuma ser aproveitada por credores não confiáveis.

Grupo de crédito residencial

O Crédito Residencial se esforça para ampliar a inclusão financeira para a população sem conta bancária, proporcionando uma experiência de empréstimo positiva e segura. Para garantir que essa população carente tenha uma experiência de empréstimo positiva, o Crédito Residencial usa uma variedade de dados alternativos - incluindo informações de telecomunicações e transacionais - para prever a capacidade de reembolso de seus clientes.

Embora a Home Credit esteja usando vários métodos estatísticos e de aprendizado de máquina para fazer essas previsões, eles estão desafiando os Kagglers a ajudá-los a desbloquear todo o potencial de seus dados. Isso garantirá que os clientes capazes de reembolsar não sejam rejeitados e que os empréstimos sejam concedidos com um principal, vencimento e calendário de reembolso que capacitará seus clientes a terem sucesso.

# Transformações

## Converter variaveis textos/categoricos para numericos/categoticos

```python
name_category_features = []
for col in category_features:
    df["CAT_" + col] = df[col].cat.codes
    name_category_features.append("CAT_" + col)


df.drop(category_features, axis=1, inplace=True)
df.head()
```

## Retirar registros que tenha variaveis com outliers (Escor Z)

```python
des_target = df["TARGET"].unique()

name_columns = []
name_columns.extend(interger_features)
name_columns.extend(float_features)
```

```python
for variavel in name_columns:
    df_default = []
    for value, target in enumerate(des_target):
        dados = df[df["TARGET"]==value]
        print(f"Classe: {des_target[value]} - variavel: {variavel} ")

        mean, std = dados[variavel].mean(), dados[variavel].std()
        min, max  = dados[variavel].min(), dados[variavel].max()
        print("Real Min: %.3f Real Max: %.3f " % (min, max))


        cut_off = std * 3
        lower, upper = mean - cut_off, mean + cut_off
        print("Limit Min: %.3f Limit Max: %.3f " % (lower, upper))

        dataset = dados[dados[variavel] >= lower]
        if len(dataset) == 0:
            dataset = dados[dados[variavel] <= upper]
        else:
            dataset = dataset[dataset[variavel] <= upper]

        df_default.append(dataset)
        
        outliers = pd.concat([dados[dados[variavel] <= lower], dados[dados[variavel] > upper]])
        print("Identificar outliers: %d \n" % len(outliers))

    df = pd.concat(df_default)


```

# Variaveis preditoras

## Identificar as melhores relações de variaveis para a variavel taget (random forest)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

caterical_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='constant')),
    ('one_hot',  OneHotEncoder())
])


preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, name_columns),
        ('cat', caterical_transformer, name_category_features)
    ]
)
```

```python
SEED = 10

pipe = Pipeline(steps=[('preprocessor', preprocessor),
                ('classifier', RandomForestClassifier(random_state=SEED, n_estimators=50))                
])

pipe.fit(X_scaler, y)

onehot_columns = list(pipe.named_steps['preprocessor'].named_transformers_['cat'].named_steps['one_hot'].get_feature_names(input_features=name_category_features))
numeric_features_list = list(name_columns)
numeric_features_list.extend(onehot_columns)
```

```python
import eli5

w = eli5.explain_weights(pipe.named_steps['classifier'], top=25, feature_names=numeric_features_list)
feature = [featureweights.feature for featureweights in w.feature_importances.importances]


diff_listy = list(set(feature) - set(name_category_features))
diff_listy = list(set(diff_listy) - set(name_columns))



for index, valor in enumerate(feature):
    if valor in diff_listy:
        feature[index] = valor[: -2]


feature = set(feature)

df_relevante = X_scaler[feature]
df_relevante.head()
```

# Aplicar o PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
pca.fit(df_relevante)

X_trans_pca = pca.fit_transform(df_relevante)
X_trans_pca
```

# Aplicar o Modelo

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

treino_x, teste_x, treino_y, teste_y = train_test_split(X_trans_pca, y, test_size=0.20, stratify=y, random_state=SEED)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))
```

## RandomForestClassifier

```python
modelo_RF = RandomForestClassifier(n_estimators=150, random_state=SEED)
modelo_RF.fit(treino_x, treino_y)
previsoes = modelo_RF.predict(teste_x)

acurracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia do modelo RandomForestClassifier foi %.2f%%" % acurracia)
```

## Teste em muitos modfelos ao mesmo tempo

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Decision Tree",
    "Neural Net",
    "Ada Boost",
    "Naive Bayes",
    "Quadratic Discriminant Analysis",
    "Bagging"
]

classitiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025, random_state=SEED),
    SVC(gamma=2, C=1, random_state=SEED),
    DecisionTreeClassifier(random_state=SEED),
    MLPClassifier(random_state=SEED),
    AdaBoostClassifier(n_estimators=150, random_state=SEED),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    BaggingClassifier(KNeighborsClassifier(3), random_state=SEED)
]
```

```python
for name, clf in zip(names, classitiers):
    clf.fit(treino_x, treino_y)
    previsoes = clf.predict(teste_x)

    acurracia = accuracy_score(teste_y, previsoes) * 100
    print("A acurácia do modelo " + name + " foi %.2f%%" % acurracia)
```
